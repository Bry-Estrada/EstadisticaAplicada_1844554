```python
## Libreria
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Datos
personas = ['Jorge', 'Raul', 'Ana', 'Lissa']
edad = [27,16,21,30]
dic1 = {'Municipio':['Escobedo','Apodaca','Guadalupe','Santiago'], 'Nacimientos':[123,345,278,321],'Muertos':[23,17,56,32],'Matrimonios': [76,45,89,12]}
datos_mun = pd.DataFrame(dic1)
```

```python
datos_mun
```

```python
personas, edad
```

```python
plt.bar(personas,edad,color='lightblue')
plt.title('Edades')
plt.xlabel('Nombres')
plt.ylabel('Edad en años')
plt.show()
```

```python
plt.bar(datos_mun['Municipio'], datos_mun['Matrimonios'], color = 'lightgreen')
plt.title('Matrimonios en Nuevo León en 20XX')
plt.xlabel('Municipios')
plt.ylabel('Numero de Matrimonios')
plt.show()
```

```python
#
#
#
fig,(graf1,graf2) = plt.subplots(1,2)
#

## Grafica 1
graf1.bar(personas,edad)
## Grafica 2
graf2.bar(datos_mun['Municipio'], datos_mun['Matrimonios'])
```

## Graficas con Pandas

Graficador de pandas esta construido con matplolib

dataframe.plot.tipodegrafica()

```python
#Grafica con pandas
datos_mun.plot.bar()
#Pandas toma por default el index como label de x
```

```python
datos_mun['Matrimonios'].plot.bar()
plt.title('Matrimonios NL')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#hacemos a los municipios como index
datos_mun = datos_mun.set_index('Municipio')

datos_mun['Matrimonios'].plot.bar()
plt.title('Matrimonios en NL')
plt.xlabel('Municipio')
plt.ylabel('Y')
plt.show()
```

```python
## Barras apiladas
datos_mun.plot.bar(stacked=True);
```

```python
## Grafica de lineas
datos_mun.plot.line();
```

```python
##Grafica de areas
datos_mun.plot.area()
plt.show()
```
