## Numpy

Sirve para manejar numeros en python

```python
import numpy as np
```

La estructura basica de numpy son los arreglos (arrays), que son una coleccion de objetos del mismo tipo (Numericos)

```python
# crear la lista con las ventas de paletas
ventas = [12,23,34,45,56]
```

```python
ventas = np.array(ventas)
```

```python
type(ventas)
```

```python
# Promedio
np.mean(ventas)
```

```python
# Suma
np.sum(ventas)
```

```python
# Max
np.max(ventas)
```

```python
# Min
np.min(ventas)
```

```python
# Varianza Poblacional 
np.var(ventas)
```

```python
# Varianza Muestral
np.var(ventas, ddof=1)
```

```python
# Desviacion estandar poblacional
np.std(ventas)
```

```python
# Desviacion muestral
np.std(ventas,ddof=1)
```
