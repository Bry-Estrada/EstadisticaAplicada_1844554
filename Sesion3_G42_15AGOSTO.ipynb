## np.select
np.select(condicion, valor condicion, valor si ninguna condicion se cumple)

```python
condiciones = [df_clase['Edad']<=17,
              df_clase['Edad']<=23,
              df_clase['Edad']>=23]
valores = ['Preparatoria','Universidad','Trabajando']
df_clase['Situacion']= np.select(condiciones,valores,default='Invalido')
```

```python
df_clase
```

```python
df_clase.count()
```

```python
df_clase['Edad'].count()
```

```python
df_clase.value_counts('Mayoria de Edad')
```

```python
df_clase.groupby('Mayoria de Edad').mean(numeric_only=True).round(1)['Edad']
```

```python
# aggregate agg
df_clase.agg('count')
```

```python
df_clase.groupby('Mayoria de Edad')['Edad'].agg(['mean','std']).round(1)
```
