## Problema 2

```python
poblacion = pd.read_csv('poblacion_Asgard.csv')
```

```python
poblacion.sort_values(['Estado','Municipio'])
```

```python
print('Hay un total de ',poblacion.value_counts('Estado').count(),' estados')
```

```python
poblacion.value_counts('Estado')
```

```python
print('Hay un total de',poblacion['Municipio'].count(),' municipios')
```

```python
print('El país tiene un total de ',poblacion['Población'].sum(),' personas')
```

```python
poblacion.groupby('Estado').sum()['Población']
```

```python
mediana = poblacion['Población'].median()
```

```python
poblacion['Clasificación'] = np.where(poblacion['Población'] >= mediana, 'Muy poblados','Poco poblados')
poblacion
```

```python
poblacion.sort_values(['Población','Municipio']).tail()
```

```python
poblacion.sort_values(['Población','Municipio']).head()
```

```python
Pobla_max = poblacion.groupby('Estado')['Población'].idxmax()
```

```python
poblacion.loc[Pobla_max]
```

```python
munpobla = poblacion.sort_values('Población').tail(15)
```

```python
munpobla_sorted = munpobla.sort_values(by='Población', ascending=False)
```

```python
plt.bar(munpobla_sorted['Municipio'],munpobla_sorted['Población'],color='red')
plt.title('15 Municipios mas poblados de Asgard')
plt.xlabel('Municipios')
plt.ylabel('Población')
plt.xticks(rotation=90)
plt.tight_layout()
plt.text(0.2, -0.1, 'La población esta \n expresada en \n millones ', transform=plt.gcf().transFigure, ha='center')
plt.show()
```

```python
peligen = peliculas['category'].value_counts()
```

```python
plt.bar(peligen.index,peligen.values,color='red')
plt.title('Peliculas por genero')
plt.xlabel('Genero')
plt.ylabel('Cantidad de peliculas')
plt.xticks(rotation=90)
plt.tight_layout()
#plt.text(0.2, -0.1, 'La poblacion esta \n esta expredada en \n millones ', transform=plt.gcf().transFigure, ha='center')
plt.show()
```
